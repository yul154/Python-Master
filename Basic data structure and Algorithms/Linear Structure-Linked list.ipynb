# Ordered-Linked list

```python
def add(self,item):
    current=self.head
    previous=None
    temp=Node(item)
    while current!=None:
        if current.value>item:
            break
        previous=current
        current=current.getNext()
    if previous==None:
        temp.setNext(self.head)
        self.head=temp
    else:
        previous.setNext(temp)
        temp.setNext(current)
```

# Insertion sort in Linked List

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def insertionSortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head:
            return head
        dummy=ListNode(0)
        dummy.next=head
        cur=head
        while cur.next:
            if cur.next.val<cur.val:
                pre=dummy
                while pre.next.val<cur.next.val:
                    pre=pre.next
                    
                tmp=cur.next
                cur.next=tmp.next
                tmp.next=pre.next
                pre.next=tmp
            else:
                cur=cur.next
        return dummy.next
```

# Merge sort in Linked List

```python
class Solution:
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head or not head.next:
            return head
        slow,fast=head,head.next
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            
        right=slow.next
        slow.next=None
        
        r=self.sortList(right)
        l=self.sortList(head)
        return self.merge(l,r)
    
    def merge(self,left,right):
        if not left or not right:
            return left or right
        if left.val>right.val:
            left,right=right,left
        dummy=pre=left
        left=left.next
        while left and right:
            if left.val< right.val:
                left=left.next
            else:
                cur=pre.next
                pre.next=right
                temp=right.next
                right.next=cur
                right=temp
            pre=pre.next
        pre.next=left or right   
        return dummy
    def merge(self,l1,l2):
        dummy=head=ListNode(None)
        while l1!=None and l2!=None:
            if l1.val<l2.val:
                dummy.next=l1
                l1=l1.next
            else:
                dummy.next=l2
                l2=l2.next
            dummy=dummy.next
        if l1!=None:
            dummy.next=l1
        if l2!=None:
            dummy.next=l2
        
        return head.next
```

```python
def isPalindrome(self, head):
    fast = slow = head
    # find the mid node
    while fast and fast.next:
        fast = fast.next.next
        slow = slow.next
    # reverse the second half
    node = None
    while slow:
        nxt = slow.next
        slow.next = node
        node = slow
        slow = nxt
    # compare the first and second half nodes
    while node: # while node and head:
        if node.val != head.val:
            return False
        node = node.next
        head = head.next
    return True
```
